```python
#handling nan values in Billing Address and then Filling data in it and storing the data of different shipping and billing address in a csv file
df['Payment Type'] = df['Payment Method'].str.split('₹').str[0].str.strip()

# Identify rows where billing address is NaN
missing_billing = df[df['Billing Street Address'].isnull()]

# Iterate over rows with missing billing address
for index, row in missing_billing.iterrows():
    # Check if payment type is 'Offline'
    if pd.notnull(row['Payment Type']) and 'Offline' in row['Payment Type']:
        # Fill billing address with shipping address
        df.at[index, 'Billing Name'] = row['Shipping Name']
        df.at[index, 'Billing Country'] = row['Shipping Country']
        df.at[index, 'Billing Street Address'] = row['Shipping Street Address']
        df.at[index, 'Billing Street Address 2'] = row['Shipping Street Address 2']
        df.at[index, 'Billing City'] = row['Shipping City']
        df.at[index, 'Billing State'] = row['Shipping State']
        df.at[index, 'Billing Zip'] = row['Shipping Zip']

different_shipping_address = df[(df['Shipping Name'] != df['Billing Name']) |
                                (df['Shipping Country'] != df['Billing Country']) |
                                (df['Shipping Street Address'] != df['Billing Street Address']) |
                                (df['Shipping Street Address 2'] != df['Billing Street Address 2']) |
                                (df['Shipping City'] != df['Billing City']) |
                                (df['Shipping State'] != df['Billing State']) |
                                (df['Shipping Zip'] != df['Billing Zip'])]

# Save the filtered rows to a CSV file
different_shipping_address.to_csv('Different_shipping_address.csv', index=False)
```

```python
#Multiple orders of same item by a single person
address_grouped = df.groupby(['Billing Street Address', 'Billing City', 'Billing State', 'Billing Zip']).apply(lambda x: x.groupby('LineItem Name').size().reset_index(name='item_count'))

# Filter out groups where any item occurs more than once
multiple_item_addresses = address_grouped[address_grouped['item_count'] > 1]
multiple_item_addresses.to_csv('multiple_item_addresses.csv', index=False)
```

```python
item_quantities = df.groupby('LineItem Name')['LineItem Qty'].sum()

# Sort the items by total quantity in descending order and take the top 20
top_20_items = item_quantities.sort_values(ascending=False).head(20)

# Plot the top 20 items
plt.figure(figsize=(10, 6))
top_20_items.plot(kind='bar', color='skyblue')
plt.title('Top 20 Highest Quantity Items')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
unusually_large_orders = df[df['LineItem Qty'] > 5]

# Save the filtered dataset as 'unusually_large_orders.csv'
unusually_large_orders.to_csv('unusually_large_orders.csv', index=False)

# Group the data by order number and sum the quantities for each item
order_item_quantities = unusually_large_orders.groupby(['Order #', 'LineItem Name'])['LineItem Qty'].sum().reset_index()

# Sort the orders by total quantity in descending order and take the top 10
top_10_large_orders = order_item_quantities.groupby('Order #').sum().nlargest(10, 'LineItem Qty')

```

```python
order_item_quantities = unusually_large_orders.groupby(['Order #', 'LineItem Name'])['LineItem Qty'].sum().reset_index()

# Sort the orders by total quantity in descending order and take the top 10
top_10_large_orders = order_item_quantities.groupby('LineItem Name').sum().nlargest(10, 'LineItem Qty')

# Plot the top 10 large orders with their item names
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_large_orders.reset_index(), y='LineItem Name', x='LineItem Qty', palette='viridis')
plt.title('Top 10 Large Orders with Item Names')
plt.xlabel('Total Quantity')
plt.ylabel('Item Name')
plt.show()
```

```python
# 4 seeing address with multiple payment method
df['Payment Type'] = df['Payment Method'].str.split('₹').str[0].str.strip()

# Group by shipping address and check for multiple unique payment types
multiple_payment_addresses = df.groupby('Shipping Street Address').apply(lambda x: x['Payment Type'].nunique() > 1)

# Extract shipping addresses with multiple unique payment types
multiple_payment_addresses = multiple_payment_addresses[multiple_payment_addresses].index
rows_with_multiple_payment_types = df[df['Shipping Street Address'].isin(multiple_payment_addresses)]
rows_with_multiple_payment_types.to_csv('rows_with_multiple_payment_types.csv', index=False)
```
